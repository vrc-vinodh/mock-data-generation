For generating mock data used a logic normal distribution,  normal distribution also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.

While generating data if we we use random sampling there is a equal chance of ocuuring event. By using normal distribution we can set % population for gender. e.g if -1 to +1 are limits set for gender male then, there is a chance that around 68% gender will be males and rest will be females.
