```python
import numpy as np
import pandas as pd
import random
from random import randint
from datetime import date
import datetime
```

```python
c_role = pd.read_csv('present_company_grade_pay.csv')
paneldata = pd.read_csv('paneldata.csv')
```

```python
paneldata.columns = ['0','id','p_start','exp_yr_count','medical_leaves']
```

```python
merge = pd.merge_ordered(c_role,paneldata,on=['id','p_start'],how='right')
paneldatayearly = merge.loc[:,['id','p_start','exp_yr_count','job_title','department','grade','salary','medical_leaves']]
paneldatayearly.head(5)
```

```python
paneldatayearly['sal1'] = paneldatayearly['salary']
paneldatayearly['sal1'] = paneldatayearly['sal1'].ffill()
paneldatayearly['job_title'] = paneldatayearly['job_title'].ffill()
paneldatayearly['department'] = paneldatayearly['department'].ffill()
paneldatayearly['grade'] = paneldatayearly['grade'].ffill()
```

```python
value = paneldatayearly['sal1'] + paneldatayearly['p_start']*paneldatayearly['exp_yr_count']
paneldatayearly['salary'] = paneldatayearly['salary'].fillna(value)
paneldatayearly.head(25)
```

```python
paneldatayearly.to_csv('panel_data_yearly_merged_salary.csv',index=False)
```
