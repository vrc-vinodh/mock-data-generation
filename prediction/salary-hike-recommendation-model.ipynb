```python
data1 = data1.drop_duplicates(subset='id', keep='last')
data1.head()
```

```python
data1['salary_difference'] = data1['salary'] - data1['avg_salary']
data1['salary_pct_dif'] = 100*data1['salary_difference']/data1['avg_salary']
data1['salary_pct_dif'] = data1['salary_pct_dif'].round(2)
data1.head()
```

```python
rating = pd.read_csv('rating.csv', index_col=0)
avg_rating = pd.read_csv('avg-rating.csv', index_col=0)

rf = pd.merge(data1,rating,on=['id','exp_yr'])
model = pd.merge(rf ,avg_rating,on='id')

model.tail()
```

```python
model.columns = ['id', 'job_title', 'exp_yr', 'exp_yr_role', 'salary', 'avg_salary',
       'salary_dif', 'pct_dif', 'year', 'month', 'Rating', 'past_3yrs_rat']
```

```python
model = model[['id', 'job_title', 'exp_yr', 'exp_yr_role', 'salary', 'avg_salary','salary_dif', 'pct_dif','Last_yr_rat',
       'past_3yrs_rat']]
```

```python
model.head()
```

1. Increase salary by more % than avg increase and give him strategic training
2. Increase salary by more % than avg increase
3. not much change in salary
4. not much change in salary, give him strategic training

mean of employees rating is 3.5

```python
#regarding salary recommendation 
model.loc[(model.pct_dif < -5) & (model.Last_yr_rat >= 3.7), 'sal_recomm'] = 'increase salary by 5%'
model.loc[(model.pct_dif < 0 ) & (model.past_3yrs_rat >= 4), 'sal_recomm'] = 'increase salary by 5'

model.loc[(model.pct_dif < -5 ) & (model.past_3yrs_rat >= 4),'sal_recomm'] = 'increase salary by 10%'
model.loc[(model.pct_dif < -10 ) & (model.Last_yr_rat >= 3.7), 'sal_recomm'] = 'increase salary by 10%'

model.loc[(model.pct_dif < -10 ) & (model.past_3yrs_rat >= 4), 'sal_recomm'] = 'increase salary by 15%'
model.loc[(model.pct_dif < -15) & (model.Last_yr_rat >= 3.7), 'sal_recomm'] = 'increase salary by 15%'
```

```python
#regarding training recommendation
model.loc[(model.Last_yr_rat <= 2.9), 'training_recomm'] = 'current training need'
model.loc[(model.Last_yr_rat >= 3.5) & (model.past_3yrs_rat >= 4), 'training_recomm'] = 'strategic training need'
```

```python
model.head(1)
```

```python
model_recommendation = model[['id','exp_yr','job_title','exp_yr_role','pct_dif','Last_yr_rat','past_3yrs_rat','sal_recomm','training_recomm']]
model_recommendation
```

```python
def recommendation_id(id):
    x = model_recommendation.loc[model_recommendation.id== id]
    return x
```
